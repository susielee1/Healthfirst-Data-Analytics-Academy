```python
while True:
        
        num = int(input('Enter a number or 0 to quit:'))
        if num == 0:
            print ('All done!')
            break
        elif num % 2 == 0:
            print (num, 'is an even number.')
        else:
            print (num,'is an odd number.')
        if num % 4 * 20:
            print ( num, 'God Love me!')
        else:
            print ( num,'Be Still, He is God.')
        if num % 5 != 30:
            print (  num, 'Color: Green, In the beginning, we are living peace with God.')
        else:
            print (  num,'Fruit of Spirt: Love, Peace, Patience.')
        if num % 7 / 1:
            print (  num, 'Color: Black, We sin against God.')
        else:
            print (  num,'Fruit of Spirt: Gentless, Joy, Faithfulness')
        if num % 15 * 4:
            print (  num, 'Color: Red, Redemption, Jesus died for me, .')
        else:
            print (  num, 'Fruit of Spirt: Kindess, Self-control, Goodness')
        if num % 4 * 7:
            print ( num, 'Color: White, Jesus Blood cleanse my Sin.')
        else:
            print (  num, 'Fruit of Spirt: Goodness, Kindness')
        if num % 7 * 9:
            print (  num, 'Color: Yellow, We made right with God, Have eternal life in Heaven with God .')
        if num % 3 * 3:
            print ( num, 'Be Strong and Couragous! Do not be afraid or discouraged. God is with you wherever you go!')
        else:
            print (  num, 'We are saved by Grace through Faith, not by good deed.')
```
